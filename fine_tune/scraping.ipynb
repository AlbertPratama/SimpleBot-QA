```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://id.wiktionary.org/wiki/Lampiran:Daftar_Swadesh_bahasa_Nias'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')

# <table class="wikitable sortable jquery-tablesorter">
# <thead>
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

## Titles

```python
word_titles = soup.find_all('th')
word_titles
```

```python
word_table_titles = [title.text.strip() for title in word_titles]

word_table_titles
```

```python
import pandas as pd

df = pd.DataFrame(columns=word_table_titles)
df = df.drop('Daftar Swadesh', axis=1)

df
```

### TD and TR

```python
columns_data = table.find_all('tr')
columns_data
```

```python
for row in columns_data:
  row_data = row.find_all('td')
  ind_row_data = [data.text.strip() for data in row_data]
  print(ind_row_data)
```

```python
print(ind_row_data)
```

```python
for rows in columns_data[1:]:
  rows_data = rows.find_all('td')
  indiv_row_data = [datas.text.strip() for datas in rows_data]
  print(indiv_row_data)

  length = len(df)
  df.loc[length] = indiv_row_data
```

```python
df
```

### Saving data into CSV

```python
df.to_csv('./dataset/kosa_kata_nias.csv', index=False)
```
